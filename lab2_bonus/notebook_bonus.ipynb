# Bonus: 双星绕转引力势场动画 (Animation)

真正的物理系统是动态的。在这个进阶挑战中，我们将使用 `matplotlib.animation` 制作一个引力势场随时间演化的动画。

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# 我们直接在这里写一个简单的势场函数，方便动画调用
def calc_V(X, Y, pos1, pos2, M1=1.0, M2=0.5, eps=0.1):
    R1 = np.sqrt((X - pos1[0])**2 + (Y - pos1[1])**2 + eps**2)
    R2 = np.sqrt((X - pos2[0])**2 + (Y - pos2[1])**2 + eps**2)
    return -M1/R1 - M2/R2
```

```python
x = np.linspace(-3, 3, 200)
y = np.linspace(-3, 3, 200)
X, Y = np.meshgrid(x, y)

fig, ax = plt.subplots(figsize=(6, 6))
ax.set_aspect('equal')

# 初始状态
V_init = calc_V(X, Y, (-1, 0), (1, 0))
contour = ax.contourf(X, Y, V_init, levels=30, cmap='viridis', vmin=-3, vmax=0)

def update(frame):
    ax.clear() # 清除上一帧
    ax.set_aspect('equal')
    
    # 简单的圆周运动
    theta = frame * np.pi / 20
    pos1 = (-np.cos(theta), -np.sin(theta))
    pos2 = (np.cos(theta), np.sin(theta))
    
    V = calc_V(X, Y, pos1, pos2)
    
    # 重新绘制
    c = ax.contourf(X, Y, V, levels=30, cmap='viridis', vmin=-3, vmax=0)
    ax.set_title(f'Binary System Frame {frame}')
    return c,

# 创建动画 (计算 40 帧)
ani = FuncAnimation(fig, update, frames=40, interval=100, blit=False)

# 在 Jupyter 中显示动画 (需要安装 ffmpeg 或使用 HTML5)
# HTML(ani.to_jshtml())
```
